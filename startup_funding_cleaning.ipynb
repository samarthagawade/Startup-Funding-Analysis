**Preprocessing(cleaning) steps :**

drop remark col

set index sno

rename col

convert amount to cr RS

date col

dropna

separate all investors

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.head(1)
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.rename(columns={'Date dd/mm/yyyy':'Date',
                   'Startup Name':'Startup',
                   'Industry Vertical':'Industry',
                   'SubVertical':'Subvertical',
                   'City  Location':'City',
                   'Investors Name':'Investors',
                   'InvestmentnType':'round',
                   'Amount in USD':'Amount'},inplace=True)
```

```python
df.head(1)
```

```python
df['Amount']=df['Amount'].fillna('0')
```

```python
df
```

```python
df['Amount'].astype('int')    #amount is in str format - convert it into int
```

```python
df['Amount']=df['Amount'].str.replace(',','')
```

```python
df['Amount'].astype('int')  #again error
```

```python
df['Amount']=df['Amount'].str.replace('undisclosed','0')
```

```python
df['Amount'].astype('int')
```

```python
df['Amount']=df['Amount'].str.replace('unknown','0')
```

```python
df['Amount'].astype('int')
```

```python
df['Amount']=df['Amount'].str.replace('Undisclosed','0')
```

```python
df['Amount'].astype('int')
```

```python
df=df[df['Amount'].str.isdigit()]
```

```python
df
```

```python
df['Amount']=df['Amount'].astype('float')
```

```python
df.info()
```

```python
def to_inr(dollar):
  inr=dollar*82.5
  return inr/10000000
```

```python
df['Amount']=df['Amount'].apply(to_inr)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'],dayfirst=True,errors='coerce')
```

```python
df.info()
```

```python
df.columns
```

```python
df.dropna(subset=['Date', 'Startup', 'Industry', 'City', 'Investors','round', 'Amount'],inplace=True)
```

```python
df.head(1)
```

```python
df.to_csv('startup_cleaned.csv',index=False)
```
